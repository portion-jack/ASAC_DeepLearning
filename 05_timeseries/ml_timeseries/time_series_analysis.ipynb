### 시계열데이터정의
1. 인덱스가 datetime형
2. 시계열데이터의 indexing, slicing 을 활용, resample 함수 사용가능

```python
data = [[1,2],[3,4],[5,8],[7,6],[9,12],[11,10],[13,16],[15,20]]
df = pd.DataFrame(data)
df.index = ['2022-01-01','2022-02-01','2022-03-01','2022-04-01',
            '2022-05-01','2022-06-01','2022-07-01','2022-08-01']

df
```

```python
df.index = pd.to_datetime(df.index, format="%Y-%m-%d")
```

```python
data = [[1,2],[3,4],[5,8],[7,6],[9,12],[11,10],[13,16],[15,20]]
df = pd.DataFrame(data)
df.index = ['2022년01월01일','2022년02월1일','2022년03월01일','2022년04월01일',
            '2022년05월01일','2022년06월01일','2022년07월01일','2022년08월01일']
```

```python
df_2014=pd.read_csv('../../99_raw_data/2014년 졸음운전 교통사고.csv',encoding='cp949')
df_2015=pd.read_csv('../../99_raw_data/2015년 졸음운전 교통사고.csv',encoding='cp949')
df_2016=pd.read_csv('../../99_raw_data/2016년 졸음운전 교통사고.csv',encoding='cp949')
```

```python
df=pd.concat([df_2014,df_2015,df_2016])
```

```python
df.set_index('구분',inplace=True)
df.index = pd.to_datetime(df.index, format='%Y년%m월')
```

```python
df.resample(rule='Q').mean()
```
