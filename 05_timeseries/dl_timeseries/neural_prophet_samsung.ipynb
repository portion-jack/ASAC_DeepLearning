```python
from neuralprophet import NeuralProphet
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import FinanceDataReader as fdr
# samsung electornics
df = fdr.DataReader('005930','2010-01-01')
df.head()
```

```python
df_main=df['Close'].reset_index()
```

```python
df_main.columns = ['ds','y']
```

```python
model = NeuralProphet()
result=model.fit(df_main)
```

```python
df_future=model.make_future_dataframe(df_main,periods=1000,n_historic_predictions=-1)
```

```python
pred_curr=model.predict(df_main)
pred_future=model.predict(df_future)
```

```python
model.plot(pd.concat([pred_curr,pred_future]),plotting_backend='plotly',figsize=(14, 10))
```

```python
df_future=model.make_future_dataframe(df_main,periods=180,n_historic_predictions=-1)
pred_future=model.predict(df_future)
model.plot(pred_future,plotting_backend='plotly',figsize=(14, 10))
```
