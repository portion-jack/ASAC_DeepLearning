```python
# prob_2
import numpy as np
import pandas as pd
df_raw=pd.read_csv('../raw_data/electric.csv',index_col=0)

x_data=df_raw['전기사용량'].values
y_data=df_raw['전기생산량'].values
```

```python
# initial_condition
w = 10
b = 10
__iter__ = 10000
lr = 0.01
n = len(x_data)

# loop 
for i in range(1,__iter__+1):
    x_hat = w*x_data + b
    cost = np.sum((x_hat-y_data)**2)/n
    
    grad_w = np.sum((x_hat-y_data)*2*x_data)/n
    grad_b = np.sum((x_hat-y_data)*2)/n
    
    w -= lr*grad_w
    b -= lr*grad_b
    
    if i%100 == 0 :
        print(f'iter_{i} => cost : {cost} / w : {w} / b : {b} ')
```
