```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt 

x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
y_data = np.array([[0],[1],[1],[0]])
```

```python
dense_1 = Dense(units=32, input_dim=2, activation='sigmoid')
dense_2 = Dense(units=1, input_dim=32, activation='sigmoid')
model = Sequential([dense_1,dense_2])
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc'])
h =model.fit(x_data,y_data,epochs=100)
```

```python
(model.predict(x_data)>0.5)+0
```

```python
modelNN = Sequential()
modelNN.add(Dense(units=32, input_dim=2, activation='sigmoid'))
modelNN.add(Dense(units=1, input_dim=32, activation='sigmoid'))
modelNN.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc'])
h =model.fit(x_data,y_data,epochs=100)
```
