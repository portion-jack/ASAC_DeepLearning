```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('../../../raw_data/softmax.txt',sep=' ',header=None)
df
"""
0,1 -> feature
2,3,4 -> target
# 가장 원시적인 classification
-> 2 binomial classification
-> 3 binomial classification
-> 4 binomial classification

"""
```

```python
x_data = df.iloc[:,:-3]
y_data = df.iloc[:,-3:]
```

```python
print(x_data.shape)
print(y_data.shape)
```

```python
x = tf.constant(x_data,tf.float32)
y = tf.constant(y_data,tf.float32)
```

```python
initX = tf.initializers.GlorotUniform()
w = tf.Variable(initX(shape = [2,3])) # 특성데이터 갯수, 라벨의 갯수
b = tf.Variable(initX(shape = [3]))   # 라벨의 갯수(분류의 갯수)
```

```python
def cost():
    z = tf.matmul(x,w) + b
    c_i = tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=z)
    c = tf.reduce_mean(c_i)
    return c

def hxFn(x_data):
    xd = tf.constant(x_data,tf.float32)
    z = tf.matmul(xd,w) + b
    hx = tf.nn.softmax(z)
    return hx.numpy().argmax(axis=1)
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(cost,var_list=[w,b])
    print(i,'cost: ', cost().numpy(), 'acc : '
          ,accuracy_score(y.numpy().argmax(axis=1),hxFn(x_data))
        )
```
