```python
import tensorflow as tf
from tensorflow.keras.optimizers import SGD

# Stochastic Gradient Descent
```

```python
x = tf.constant(value=[1,2,3], dtype=tf.float32)
y = tf.constant(value=[1,2,3], dtype=tf.float32)
```

```python
w = tf.Variable(10.0)
b = tf.Variable(10.0)
```

```python
def cost():
    x_hat = w*x + b
    c = tf.reduce_mean((x_hat-y)**2)
    return c
```

```python
optimizer = SGD(learning_rate=0.01)
for i in range(2000):
    optimizer.minimize( cost, var_list = [w,b])
    
    """
    -- optimizer.minimize --
    
    w = w-0.01 * w_derivative 
    w.assign_sub(0.01*w_derivative)
    
    b = b-0.01 * b_derivative 
    b.assign_sub(0.01*b_derivative)
    """
    
    print(f'iter_{i+1} : cost : {cost().numpy()}')
```
