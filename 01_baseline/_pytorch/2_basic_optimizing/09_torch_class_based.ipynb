````python
class XORModel(Module):
    """
    
    XORModel
    2 -> 1
        (nn1)                 (nn2)
    (n,2) -> (n,16) -> sigmoid -> (n,1) -> sigmoid
    
    """
    def __init__(self) -> None:
        super().__init__()
        
        self.nn1 = Linear(2,16) # w1 : (2,16) b = (16) => 
        self.sig1 = Sigmoid()
        
        self.nn2 = Linear(16,1) # w1 : (16,1) b = (1)
        self.sig2 = Sigmoid()
    
    def forward(self, x):
        """
        1. self.nn1.forward(x) <=> matmul(w1,x) + b1
        2. self.sig1.forward(x) <=> 1/(1+exp(-z))
        3. self.nn2.forward(x) <=> matmul(w2,x) + b2
        4. self.sig2.forward(x) <=> 1/(1+exp(-z))
        """
        print('forward call')
        x = self.nn1(x) # 1
        x = self.sig1(x) # 2 
        x = self.nn2(x) # 3
        x = self.sig2(x) # 4
        return x
````

````python
# Data
x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
x = torch.FloatTensor(x_data)

y_data = np.array([[0],[1],[1],[0]])
y = torch.FloatTensor(y_data)
````

````python
model = XORModel()
loss_fn = BCELoss()
optimizer = Adam(model.parameters(), lr=0.1)

for epoch in range(200):
    optimizer.zero_grad()
    
    x_hat = model(x) # model.forward(x)
    
    cost = loss_fn(x_hat,y)
    cost.backward()
    optimizer.step()
    print(f'epoch => {epoch} / cost => {cost.item()}')
````
