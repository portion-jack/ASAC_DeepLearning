```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential, BCELoss, Sigmoid

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
y_data = np.array([[0],[1],[1],[0]])
```

```python
x = torch.FloatTensor(x_data)
y = torch.FloatTensor(y_data)
```

```python
model = Sequential()
model.add_module('lr_1',Linear(2,16))
model.add_module('sig_1',Sigmoid())
"""
For TF :
dense = Dense(units=16, input_dim=2, activation='sigmoid')
=> activation available 
"""
model.add_module('lr_2',Linear(16,1))
model.add_module('sig_2',Sigmoid())

loss_fn = BCELoss()
optimizer = Adam(model.parameters(), lr = 0.1)

for epoch in range(200):
    optimizer.zero_grad()
    hx = model(x)
    """
    z1 = matmult(x,w1) + b1
    hx_1 = sigmoid(z1)
    z2 = matmult(hx_1,w2) + b2
    hx_2 = sigmoid(z2)
    return hx_2
    """
    c = loss_fn(hx,y)
    c.backward()
    optimizer.step()
    acc = accuracy_score(y_data,((hx>0.5)+0).detach().numpy())
    print(f'epoch => {epoch} / cost => {c.item()} / acc => {acc}')
```
