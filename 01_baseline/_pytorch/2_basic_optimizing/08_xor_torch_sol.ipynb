```python
import torch
from torch.optim import Adam
import torch.nn.functional as F

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
y_data = np.array([[0],[1],[1],[0]])
```

```python
x = torch.FloatTensor(x_data)
y = torch.FloatTensor(y_data)
```

```python
w1=torch.empty([2,16], requires_grad=True)
w2=torch.empty([16,1], requires_grad=True)
torch.nn.init.uniform_(w1)
torch.nn.init.uniform_(w2)

b1=torch.empty([16], requires_grad=True)
b2=torch.empty([1], requires_grad=True)
torch.nn.init.uniform_(b1)
torch.nn.init.uniform_(b2)
```

```python
def cost():
    z1 = torch.matmul(x,w1) + b1
    hx1 = torch.sigmoid(z1)
    z2 = torch.matmul(hx1,w2) + b2
    c_i = F.binary_cross_entropy_with_logits(z2,y)
    c = torch.mean(c_i)
    return c

"""

F.binary_cross_entropy => without sigmoid
F.binary_cross_entropy_with_logits = > with sigmoid

"""
```

```python
def hxFn(x_data):
    xd = torch.FloatTensor(x_data)
    z1 = torch.matmul(xd,w1) + b1
    x1 = torch.sigmoid(z1)
    z2 = torch.matmul(x1,w2) + b2
    hx = torch.sigmoid(z2)
    return (hx.detach().numpy()>0.5) + 0
```

```python
optimizer = Adam([w1,w2,b1,b2], lr=0.1)
for epoch in range(1000):
    c = cost()
    optimizer.zero_grad()
    c.backward()
    optimizer.step()
    c_acc = accuracy_score(hxFn(x_data),y_data)
    print(f'cost = > {c.item()} acc => {c_acc}')
```
