```python
import torch
from torch.optim import Adam
from torch.nn import Linear, Sequential
from torch.nn import Softmax, CrossEntropyLoss

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('./raw_data/softmax.txt',sep=' ',header=None)
df
```

```python
x_data = df.iloc[:,:-3]
y_data = df.iloc[:,-3:]
```

```python
x = torch.FloatTensor(x_data.values)
y = torch.FloatTensor(y_data.values)
```

```python
model = Sequential()
model.add_module('nn',Linear(2,3))
model.add_module('soft1',Softmax(dim=1))
loss_fn = CrossEntropyLoss()
optimizer = Adam(model.parameters(),lr=0.1)

for _iter_ in range(3000):
    optimizer.zero_grad()
    hx = model.forward(x)
    # hx = model(x)
    c = loss_fn(hx,y)
    c.backward()
    optimizer.step()
    acc = accuracy_score(model.forward(x).detach().numpy().argmax(axis=1),
                         y.detach().numpy().argmax(axis=1))
    print(f'_iter_ => {_iter_} / cost => {c.item()} / acc => {acc}') 
```
