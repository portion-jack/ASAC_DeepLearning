```python
rss_khan = "https://www.khan.co.kr/rss/rssdata/total_news.xml"
feed = feedparser.parse(rss_khan)
```

```python
links = [n['link'] for n in feed['entries']]
```

```python
klist = list()

for ln in links[:3]:
    article = newspaper.Article(ln)
    article.download()
    article.parse()
    print(article.text)
    klist.append(article.text)
    print(ln)
    print('==')
```
