```python
import matplotlib.pyplot as plt
import numpy as np
x_data = np.array([1,2,3,4,5])
y_data = np.array([3,5,7,9,11])
```

```python
plt.xlim([0,15])
plt.ylim([0,15])
plt.scatter(x_data,y_data)
plt.show()
```

```python
w = 10
b = 10
epochs = 2000
learning_rate = 1e-2
n = len(x_data)
```

```python
c_hist = list()
for i in range(epochs):
    x_hat = w*x_data + b
    cost = np.sum( (x_hat-y_data)**2)/n
    gradient_w = np.sum( ((w*x_data + b)-y_data)*2*x_data)/n
    gradient_b = np.sum( ((w*x_data + b)-y_data)*2)/n
    w = w-learning_rate*gradient_w
    b = b-learning_rate*gradient_b    
    print(f'cost : {cost} w : {w} b : {b}')
    c_hist.append(cost)
```

```python
print(f'최종 w: {w} 최종 b: {b}')
```
