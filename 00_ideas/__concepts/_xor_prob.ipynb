```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
```

```python
x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
y_data = np.array([[0],[1],[1],[0]])
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
initX = tf.initializers.GlorotUniform(seed=123)
w = tf.Variable(initX(shape=[2,1]))
b = tf.Variable(initX(shape=[1]))
```

```python
def cost():
    z = tf.matmul(x,w) + b
    c_i = tf.nn.sigmoid_cross_entropy_with_logits(labels=y,logits=z)
    c = tf.reduce_mean(c_i)
    return c
```

```python
optimzier = Adam(learning_rate=0.1)
for i in range(1000):
    optimzier.minimize(cost, var_list=[w,b])
    print(i, 'cost',cost().numpy())
```

```python
def hxFn(x_data):
    xd = tf.constant(x_data,tf.float32)
    z = tf.matmul(xd,w) + b
    hx = tf.nn.sigmoid(z)
    return (hx.numpy()>0.5) + 0
```
