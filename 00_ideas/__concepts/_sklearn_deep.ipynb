```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
y_data = np.array([[0],[1],[1],[0]])
```

```python
model_logistic = LogisticRegression(max_iter=100,verbose=1)
model_logistic.fit(x_data,y_data.reshape(-1))
```

```python
model_logistic.predict(x_data)
```

```python
model_logistic.score(x_data,y_data)
```

```python
print(model_logistic.coef_)
print(model_logistic.intercept_)
```

```python
"""
Multi - Layer - Perceptron - Classifier
"""
model_mlp = MLPClassifier(verbose=1,
                          hidden_layer_sizes=(100,), # default
                          learning_rate='adaptive',
                          max_iter=1000)
model_mlp.fit(x_data,y_data.reshape(-1))
```

```python
model_mlp.predict(x_data)
```

```python
model_mlp.score(x_data,y_data)
```

```python
model_mlp.get_params()
```

```python
print(len(model_mlp.coefs_))
print(len(model_mlp.intercepts_))
```

```python
print(model_mlp.coefs_[0].shape)
print(model_mlp.intercepts_[0].shape)
```

```python
print(model_mlp.coefs_[1].shape)
print(model_mlp.intercepts_[1].shape)
```

```python
model_mlp = MLPClassifier(verbose=0,
                          hidden_layer_sizes=(100,50),
                          learning_rate='adaptive',
                          max_iter=30)
model_mlp.fit(x_data,y_data.reshape(-1))

print(len(model_mlp.coefs_))
print(len(model_mlp.intercepts_))
print(model_mlp.coefs_[0].shape)
print(model_mlp.intercepts_[0].shape)
print(model_mlp.coefs_[1].shape)
print(model_mlp.intercepts_[1].shape)
print(model_mlp.coefs_[2].shape)
print(model_mlp.intercepts_[2].shape)
print(model_mlp.coefs_[3].shape)
print(model_mlp.intercepts_[3].shape)
```

```python
model_mlp = MLPClassifier(verbose=0,
                          hidden_layer_sizes=(100,50,30),
                          learning_rate='adaptive',
                          max_iter=30)
model_mlp.fit(x_data,y_data.reshape(-1))

print(len(model_mlp.coefs_))
print(len(model_mlp.intercepts_))
print(model_mlp.coefs_[0].shape)
print(model_mlp.intercepts_[0].shape)
print(model_mlp.coefs_[1].shape)
print(model_mlp.intercepts_[1].shape)
print(model_mlp.coefs_[2].shape)
print(model_mlp.intercepts_[2].shape)
print(model_mlp.coefs_[3].shape)
print(model_mlp.intercepts_[3].shape)
```
