### Classification

```
Binary Classification : Logistic Regression
Multi Classification : Soft-max
```

````python
import math
import matplotlib.pyplot as plt
import numpy as np
````

````python
def sigmoid(z):
    return 1 / (1+math.e**-z)

def sigmoid(z):
    return 1 / (1+math.exp(-z))
````

````python
print(sigmoid(-10))
print(sigmoid(-1))
print(sigmoid(0))
print(sigmoid(1))
print(sigmoid(10))
````

````python
for z in np.linspace(-10,10,50):
    s= sigmoid(z)
    plt.plot(z,s,'r+')
````
