```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
```

### MSE = cost
### Mean_Squared_Error = Cost

```python
def cost(x,y,w):
    c = 0 
    for i in range(len(x)):
        hx = w*x[i]
        c = c+(hx - y[i])**2
    return c/len(x)
```

```python
print(cost(x_data,y_data,-1))
print(cost(x_data,y_data,0))
print(cost(x_data,y_data,1))
print(cost(x_data,y_data,2))
```

```python
plt.scatter(x_data,y_data)
plt.xlim([0,5])
plt.ylim([0,5])
plt.show()
```

```python
# start end counts
np.linspace(1,10,5)
```

```python
for w in np.linspace(-3,5,50):
    c = cost(x_data,y_data,w)
    print(f'w={w} -> cost={c}')
    plt.plot(w,c,'ro')
plt.xlabel('weight')
plt.ylabel('cost')
plt.tight_layout()
plt.show()
```
