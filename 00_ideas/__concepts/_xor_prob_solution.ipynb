```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
```

```python
x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
y_data = np.array([[0],[1],[1],[0]])
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
initX = tf.initializers.GlorotUniform()
w1 = tf.Variable(initX(shape=[2,32]))
b1 = tf.Variable(initX(shape=[32]))

w2 = tf.Variable(initX(shape=[32,1]))
b2 = tf.Variable(initX(shape=[1]))
```

```python
def cost():
    z1 = tf.matmul(x,w1) + b1
    hx1 = tf.nn.sigmoid(z1)
    z = tf.matmul(hx1,w2) + b2
    
    c_i = tf.nn.sigmoid_cross_entropy_with_logits(labels=y,logits=z)
    c = tf.reduce_mean(c_i)
    return c
```

```python
optimzier = Adam(learning_rate=0.1)
for i in range(1000):
    optimzier.minimize(cost, var_list=[w1,w2,b1,b2])
    print(i, 'cost',cost().numpy())
```

```python
def hxFn(x_data):
    xd = tf.constant(x_data,tf.float32)
    z1 = tf.matmul(xd,w1) + b1
    hx1 = tf.nn.sigmoid(z1)
    z = tf.matmul(hx1,w2) + b2
    hx = tf.nn.sigmoid(z)
    return (hx.numpy()>0.5) + 0
```

```python
hxFn(x)
```

```python
initX = tf.initializers.GlorotUniform()
w1 = tf.Variable(initX(shape=[2,64]))
b1 = tf.Variable(initX(shape=[64]))

w2 = tf.Variable(initX(shape=[64,32]))
b2 = tf.Variable(initX(shape=[32]))

w3 = tf.Variable(initX(shape=[32,1]))
b3 = tf.Variable(initX(shape=[1]))

def cost():
    z1 = tf.matmul(x,w1) + b1
    hx1 = tf.nn.sigmoid(z1)
    z2 = tf.matmul(hx1,w2) + b2
    hx2 = tf.nn.sigmoid(z2)
    z = tf.matmul(hx2,w3) + b3
    c_i = tf.nn.sigmoid_cross_entropy_with_logits(labels=y,logits=z)
    c = tf.reduce_mean(c_i)
    return c

def hxFn(x_data):
    xd = tf.constant(x_data,tf.float32)
    z1 = tf.matmul(xd,w1) + b1
    hx1 = tf.nn.sigmoid(z1)
    
    z2 = tf.matmul(hx1,w2) + b2
    hx2 = tf.nn.sigmoid(z2)
    
    z = tf.matmul(hx2,w3) + b3
    hx = tf.nn.sigmoid(z)
    return (hx.numpy()>0.5)+0
```

```python
optimzier = Adam(learning_rate=0.1)
for i in range(1000):
    optimzier.minimize(cost, var_list=[w1,w2,b1,b2])
    print(i, 'cost',cost().numpy())
```
