```python
import os
import time

import urllib
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import bs4
```

```python
keywords = ['스트릿패션','츄리닝패션']
```

```python
keywords = ['정장패션']
for keyword in keywords:
    if os.path.exists(f'./img_data/{keyword}'):
        pass
    else:
        os.mkdir(f'./img_data/{keyword}')
        
    d_path = r"/Users/jack/utils/chromedriver"
    s = Service(d_path)

    driver = webdriver.Chrome(service=s)
    url = "https://www.google.com/imghp"
    driver.get(url)
    driver.find_element(By.NAME, "q").send_keys(keyword + Keys.ENTER)
    time.sleep(3)
    # 몇백개의 키워드를 받기 위해서 스크롤 내림
    print(keyword +' 키워드 스크롤 중 .............')
    elem =  driver.find_element(By.TAG_NAME, "body")
    for i in range(30):
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.1)
        
    try:
        driver.find_element(By.XPATH, '//*[@id="islmp"]/div/div/div/div/div[1]/div[2]/div[2]/input').click()
        for i in range(30):
            elem.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.1)
    except:
        pass
    keyword_links=[]
    images = driver.find_elements(By.CSS_SELECTOR, "img.rg_i.Q4LuWd") # 구글 class값이 공통적으로 rg_i.Q4LuWd
    for image in images:
        if image.get_attribute('src')!=None:
            keyword_links.append(image.get_attribute('src'))
        if len(keyword_links) >= 100:
            break

    print(keyword+' 찾은 이미지 개수 :',len(keyword_links))
    time.sleep(2)

    for num, url in enumerate(keyword_links):
        url = url
        start = time.time()
        urllib.request.urlretrieve(url, f"./img_data/{keyword}/"+keyword+"_"+str(num+1)+".jpg")
        print(str(num+1)+'/'+str(len(keyword_links))+' '+keyword+' 저장 중.... time : ' + str(time.time()-start)[:5])
    print('---모든 파일 저장 완료---')

    driver.close()
    time.sleep(3)
```
